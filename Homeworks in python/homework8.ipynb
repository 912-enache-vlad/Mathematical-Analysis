```python
from matplotlib import pyplot as plt

p = 1.25

def give_yk(xk:int):
    return -(1 - abs(xk)**p)**(1/p), (1 - abs(xk)**p)**(1/p)

ax = plt.subplot()
x_change = 1/10000

x = []          # horizontal line
y = []          # vertical line

xk = -1
yk1, yk2 = give_yk(xk)

while xk < 1:
    x.append(xk)
    y.append(yk1)
    x.append(xk)
    y.append(yk2)
    xk += x_change
    yk1, yk2 = give_yk(xk)


ax.plot(x, y, marker = ",", linestyle = "", color = "r")
ax.grid(True)
plt.gca().set_aspect("equal")
plt.show()

```
